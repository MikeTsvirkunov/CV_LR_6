```python
import json
import torch
import inspect
import re
import matplotlib.pyplot as plt
import numpy as np
```

```python
vgg_transform_scheduler_1 = json.load(open('/content/drive/MyDrive/cv6/vgg__transfom_scheduler(1).json'))
rn50_1 = json.load(open('/content/drive/MyDrive/cv6/rn50(1).json'))
resnet34__transfom_1 = json.load(open('/content/drive/MyDrive/cv6/resnet34__transfom(1).json'))
unet_4 = json.load(open('/content/drive/MyDrive/cv6/unet(4).json'))
```

```python
def debugPrint(x):
    frame = inspect.currentframe().f_back
    s = inspect.getframeinfo(frame).code_context[0]
    r = re.search(r"\((.*)\)", s).group(1)
    return r


def plot_train_valid(d: dict, model_name: str):
  xv = np.arange(0, len(d['train_losses']), len(d['train_losses']) / len(d['test_losses']))
  plt.plot(d['train_losses'], label='train_losses')
  plt.plot(xv, d['test_losses'], label='test_losses')
  plt.legend()
  plt.title(model_name.split(',')[0])
```

```python
plot_train_valid(vgg_transform_scheduler_1, debugPrint(vgg_transform_scheduler_1))
```

```python
plot_train_valid(resnet34__transfom_1, debugPrint(resnet34__transfom_1))
```

```python
plot_train_valid(rn50_1, debugPrint(rn50_1))
```

```python
plot_train_valid(unet_4, debugPrint(unet_4))
```
